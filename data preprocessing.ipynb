```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_csv_data.copy()
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df.info()
```

### drop ID

```python
df = df.drop(['ID'],axis = 1)
```

### reason for absence

```python
df['Reason for Absence'].unique()
```

```python
reason_coulums = pd.get_dummies(df['Reason for Absence'])
reason_coulums
```

```python
reason_coulums['check'] = reason_coulums.sum(axis=1)
reason_coulums
```

```python
reason_coulums['check'].sum(axis=0)
```

```python
reason_coulums['check'].unique()
```

```python
reason_coulums = reason_coulums.drop(['check'],axis = 1)
```

```python
reason_coulums = pd.get_dummies(df['Reason for Absence'],drop_first = True)
```

```python
df = df.drop(['Reason for Absence'],axis = 1)
```

### classifying the dummy variables

```python
reason_type_1 = reason_coulums.loc[:, 1:14].max(axis = 1)
reason_type_2 = reason_coulums.loc[:, 15:17].max(axis = 1)
reason_type_3 = reason_coulums.loc[:, 18:21].max(axis = 1)
reason_type_4 = reason_coulums.loc[:, 22:].max(axis = 1)
```

### concatenate the column values

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4],axis=1)
df
```

```python
df.columns.values
```

```python
coloumn_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'reason_type_1', 'reason_type_2', 'reason_type_3', 'reason_type_4']
df.columns = coloumn_names
```

### reorder columns

```python
column_names_reordered = ['reason_type_1', 'reason_type_2', 'reason_type_3', 'reason_type_4', 'Date', 'Transportation Expense',
                          'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education',
                          'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
df.head()
```

```python
df_reason_mod = df.copy()
```

### 'Date':

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
type(df_reason_mod['Date'][0])
```

### Extract the moth value:

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
list_months
```

```python
df_reason_mod.shape
```

### Extract the day of the week

```python
df_reason_mod['Date'][699].weekday()
```

```python
df_reason_mod['Date'][699]
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
```
